## Random variates

### 随机变量

```python
from __future__ import print_function, division

%matplotlib inline

import numpy as np

import nsfg
import first
import analytic

import thinkstats2
import thinkplot

import nothinkstats2
import nothinkplot
```

```python
import random
```

```python
def expovariate(lam):
    p = random.random()
    z = 1-p
    c = np.log(z)
    y = -c
    x = y/lam
    print("p:", p, " 1-p:", z, " np.log(1-p):", '\n', c, " -np.log(1-p):", y, " x:", x)
    print('\n')
    return x
```

```python
t = [expovariate(lam=2) for _ in range(1000)]
t
```

```python
cdf = thinkstats2.Cdf(t, label='Random')

thinkplot.PrePlot(2)

thinkplot.Cdf(cdf, complement=True)


thinkplot.Cdf(cdf)

thinkplot.Config(xlabel='Exponential variate', ylabel='CCDF', yscale='log')
```
