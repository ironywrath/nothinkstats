## Scatter plots

```python
from __future__ import print_function, division

%matplotlib inline

import numpy as np

import brfss

import thinkstats2
import thinkplot
```

```python
df = brfss.ReadBrfss(nrows=None)
```

```python
def SampleRows(df, nrows, replace=False):
    indices = np.random.choice(df.index, nrows, replace=replace)
    sample = df.loc[indices]
    return sample
```

```python
sample = SampleRows(df, 5000)
heights, weights = sample.htm3 , sample.wtkg2
sample.head()
```

```python
thinkplot.Scatter(heights, weights, alpha=1)
thinkplot.Config(xlabel='Height', ylabel='Weight',axis=[140,210,20,200],legend=False)
```

```python
def Jitter(values, jitter=0.5):
    n = len(values)
    return np.random.normal(0, jitter, n) + values
```

```python
heights = Jitter(heights, 1.4)
weights = Jitter(weights, 0.5)
```

```python
thinkplot.Scatter(heights, weights, alpha=1.0)
thinkplot.Config(xlabel='Height', ylabel='Weight', axis=[140,210,20,200], legend=False)
```

```python
thinkplot.Scatter(heights, weights, alpha=0.1, s=10)
thinkplot.Config(xlabel='Height', ylabel='Weight', axis=[140,210,20,200], legend=False)
```
